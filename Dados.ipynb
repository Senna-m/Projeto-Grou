```python
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)
```

```python
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df_limpo.rename(columns=novos_nomes, inplace=True)
```

```python
df_limpo.head()
```

```python
df['senioridade'].value_counts()
```

```python

senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df_limpo['senioridade'] = df_limpo['senioridade'].replace(senioridade)
df_limpo['senioridade'].value_counts()
     
```

```python

contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df_limpo['contrato'] = df_limpo['contrato'].replace(contrato)
df_limpo['contrato'].value_counts()
```

```python

tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M':	'media'

}
df_limpo['tamanho_empresa'] = df_limpo['tamanho_empresa'].replace(tamanho_empresa)
df_limpo['tamanho_empresa'].value_counts()
```

```python

mapa_trabalho = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}

df_limpo['remoto'] = df_limpo['remoto'].replace(mapa_trabalho)
df_limpo['remoto'].value_counts()
```

```python
ordem =df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).index
```

```python

plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', estimator='mean', order=ordem)
plt.title('Salário médio por nível de senioridade')
plt.ylabel('Salário Médio (USD)')
plt.xlabel('Senioridade')
plt.show()
#orde -> o parametro feito la em cima para ordenar o grafico
```
